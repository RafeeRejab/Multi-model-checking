## Plotting the MAE for all models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
MAE=pd.read_csv('XGBoost.csv')
MAE['ET']=pd.read_csv('Extra_T.csv')['Extra_T']
MAE['LR']=pd.read_csv('Linear_reg.csv')['Linear_reg']
MAE['RIG']=pd.read_csv('Ridge.csv')['Ridge']
MAE['RIGCV']=pd.read_csv('Ridgecv.csv')['Ridgecv']
MAE['LAS']=pd.read_csv('Lasso.csv')['Lasso']
del MAE['Unnamed: 0']
```

```python
plt.figure(figsize=(15,5))
for ME in MAE:
    plt.plot(MAE[ME],label=ME)
    
plt.xlabel('Model')
plt.ylabel('MAE')
plt.legend()
plt.grid()
plt.show()
```
